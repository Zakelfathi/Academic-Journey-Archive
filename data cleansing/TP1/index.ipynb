

| we can see that the columns containing prices are strings with the comma and dollar sign   |
|-----------------------------------------|

```python
# remove dollar sign and comma from price columns and convert to numerical data
def format_price_columns(dataframe):
    price_columns = ['PRICE', 'WEEKLY_PRICE', 'MONTHLY_PRICE', 'SECURITY_DEPOSIT', 'CLEANING_FEE', 'EXTRA_PEOPLE']
    for col in price_columns:
        dataframe[col] = dataframe[col].apply(lambda x: float(x.replace(',', '').replace('$', '')) if isinstance(x, str) else x)
    return dataframe
```

```python
df_boston = format_price_columns(df_boston) 
df_cambridge = format_price_columns(df_cambridge)
```

# One-hot encoded

```python
# specify the columns to be one-hot encoded
cols_to_encode = ['HOST_RESPONSE_TIME', 'HOST_RESPONSE_RATE', 'HOST_IS_SUPERHOST',
                  'HOST_HAS_PROFILE_PIC', 'HOST_IDENTITY_VERIFIED', 'AMENITIES',
                  'FIRST_REVIEW', 'LAST_REVIEW', 'INSTANT_BOOKABLE',
                  'IS_BUSINESS_TRAVEL_READY']

# perform one-hot encoding
categorical_columns_b = pd.get_dummies(categorical_columns_b, columns=cols_to_encode)

# view the resulting dataframe
categorical_columns_b.columns
```

```python
# specify the columns to be one-hot encoded
cols_to_encode = ['HOST_RESPONSE_TIME', 'HOST_RESPONSE_RATE', 'HOST_IS_SUPERHOST',
                  'HOST_HAS_PROFILE_PIC', 'HOST_IDENTITY_VERIFIED', 'AMENITIES',
                  'FIRST_REVIEW', 'LAST_REVIEW', 'INSTANT_BOOKABLE',
                  'IS_BUSINESS_TRAVEL_READY']

# perform one-hot encoding
categorical_columns_c = pd.get_dummies(categorical_columns_c, columns=cols_to_encode)

# view the resulting dataframe
categorical_columns_c.columns
```
