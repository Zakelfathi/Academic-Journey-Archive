### Map : Confirmed, Deaths

```python
df_country_wise_latest = pd.read_csv('./datasets/country_wise_latest.csv')
df_country_wise_latest
```

```python
# !pip install plotly-express
```

```python
import plotly.express as px
# group the data by country/region and sum the deaths and confirmed cases
grouped_df = df_country_wise_latest.groupby(['Country/Region']).agg({'Deaths': 'sum', 'Confirmed': 'sum'}).reset_index()

# create the map using plotly express
fig = px.choropleth(grouped_df, locations='Country/Region', locationmode='country names', color='Deaths',
                    title='Deaths by Country/Region')
fig.show()

fig = px.choropleth(grouped_df, locations='Country/Region', locationmode='country names', color='Confirmed',
                    title='Confirmed Cases by Country/Region')
fig.show()
```
